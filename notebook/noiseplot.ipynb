```r
library(R.matlab)
signal8PSK <- readMat("frame8PSK001.mat")
values8PSK <- signal8PSK['frame']$frame
px8PSK = Mod(values8PSK) * cos(Arg(values8PSK)) 
py8PSK = Mod(values8PSK) * sin(Arg(values8PSK))

signal16QAM <- readMat("frame16QAM001.mat")
values16QAM <- signal16QAM['frame']$frame
px16QAM = Mod(values16QAM) * cos(Arg(values16QAM)) 
py16QAM = Mod(values16QAM) * sin(Arg(values16QAM))

signal64QAM <- readMat("frame64QAM001.mat")
values64QAM <- signal64QAM['frame']$frame
px64QAM = Mod(values64QAM) * cos(Arg(values64QAM)) 
py64QAM = Mod(values64QAM) * sin(Arg(values64QAM))

signalBPSK <- readMat("frameBPSK001.mat")
valuesBPSK <- signalBPSK['frame']$frame
pxBPSK = Mod(valuesBPSK) * cos(Arg(valuesBPSK)) 
pyBPSK = Mod(valuesBPSK) * sin(Arg(valuesBPSK))

signalCPFSK <- readMat("frameCPFSK001.mat")
valuesCPFSK <- signalCPFSK['frame']$frame
pxCPFSK = Mod(valuesCPFSK) * cos(Arg(valuesCPFSK)) 
pyCPFSK = Mod(valuesCPFSK) * sin(Arg(valuesCPFSK))

signalGFSK <- readMat("frameGFSK001.mat")
valuesGFSK <- signalGFSK['frame']$frame
pxGFSK = Mod(valuesGFSK) * cos(Arg(valuesGFSK)) 
pyGFSK = Mod(valuesGFSK) * sin(Arg(valuesGFSK))

signalPAM4 <- readMat("framePAM4001.mat")
valuesPAM4 <- signalPAM4['frame']$frame
pxPAM4 = Mod(valuesPAM4) * cos(Arg(valuesPAM4)) 
pyPAM4 = Mod(valuesPAM4) * sin(Arg(valuesPAM4))

signalQPSK <- readMat("frameQPSK001.mat")
valuesQPSK <- signalQPSK['frame']$frame
pxQPSK = Mod(valuesQPSK) * cos(Arg(valuesQPSK)) 
pyQPSK = Mod(valuesQPSK) * sin(Arg(valuesQPSK))
```

```r
png("layout1.png", bg = "transparent")
attach(mtcars)
par(mfrow=c(2,2), oma=c(1,1,1,1), mar=c(2,2,1,1))

plot(px8PSK, py8PSK, pch=16,cex=0.4)
mtext("8PSK",side=1,line=2)
plot(px16QAM, py16QAM, pch=16,cex=0.4);
mtext("16QAM",side=1,line=2)
plot(px64QAM, py64QAM, pch=16,cex=0.4)
mtext("64QAM",side=1,line=2)
plot(pxBPSK, pyBPSK, pch=16,cex=0.4)
mtext("BPSK",side=1,line=2)

mtext(text="I",side=1,line=0,outer=TRUE)
mtext(text="Q",side=2,line=0,outer=TRUE)
dev.off()
```
